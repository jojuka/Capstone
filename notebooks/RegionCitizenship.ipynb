```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy

x_labels = [2017, 2018, 2019]
yAfrica = [59127, 60308, 61569]
yAsia = [1349196, 1395309, 1376835]
yEurope = [423299, 421962, 420541]
yNAmerica = [505948, 482505, 451746]
yOceania = [25649, 25535, 25800]
ySAmerica = [161360, 176216, 184466]
yUnknown = [9777, 7390, 6698]

plt.plot(yAfrica, color='green')
plt.plot(yAsia, color='red')
plt.plot(yEurope, color='blue')
plt.plot(yNAmerica, color='orange')
plt.plot(yOceania, color='cyan')
plt.plot(ySAmerica, color='black')
plt.plot(yUnknown, color='yellow')

plt.title('Test')

plt.xlabel('Year')
plt.ylabel('Number of Admissions')
plt.xticks(numpy.arange(len(x_labels)), x_labels, rotation=90)

green_patch=mpatches.Patch(label='Africa', color='green')
red_patch=mpatches.Patch(label='Asia', color='red')
blue_patch=mpatches.Patch(label='Europe', color='blue')
orange_patch=mpatches.Patch(label='NAmerica', color='orange')
cyan_patch=mpatches.Patch(label='Oceania', color='cyan')
black_patch=mpatches.Patch(label='SAmerica', color='black')
yellow_patch=mpatches.Patch(label='Unknown', color='yellow')
plt.legend(handles=[green_patch, red_patch, blue_patch, orange_patch, cyan_patch, black_patch, yellow_patch])

#plt.ticklabel_format(style='plain') #to remove scientific chart notation

plt.grid(True)

plt.show()
```
