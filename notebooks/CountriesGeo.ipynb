```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\jojuk\Documents\git\projects\Capstone\data\SQLCountryInfo.csv")
```

```python
df.head()
```

```python
#df.dropna(axis=1, how= 'all', inplace=False)
df.dropna(how='all', axis=1, inplace=True)
```

```python
df
```

```python
df['Longitude'].tolist()
```

```python
df['Latitude'].tolist()
```

```python
df['Coordinates'] = list(zip(df.Longitude, df.Latitude))
```

```python
df['Coordinates'] = df['Coordinates'].apply(Point)
```

```python
gdf = gpd.GeoDataFrame(df, geometry='Coordinates')
```

```python
print(gdf.head())
```

```python
world = gpd.read_file('../data/world.json')
```

```python
ax=world.plot(figsize=(20,16) )#color='8BB5D9'
gdf.plot(ax=ax, color='#75AB37')

ax.set_title("Countries of Origin for Students and Exchange Visas to U.S., 2017 to 2019")
plt.show()

plt.savefig('Countries of Origin.png')
```
